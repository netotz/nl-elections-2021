# Demostración de análisis de datos de Twitter sobre la candidata Clara Luz Flores

A continuación se obtendrá la información de todos los tweets hechos por la cuenta `@claraluzflores` comprendidos desde el 5 de marzo del 2021, fecha en la que comenzó la campaña electoral, hasta el 8 de abril del 2021, día anterior al momento de realizar el presente.

```python
from constants import CLARA_LUZ, CAMPAIGN_START
from twitter import get_tweets_by_candidate

tweets_clf = get_tweets_by_candidate(CLARA_LUZ, since=CAMPAIGN_START)
tweets_clf
```

La gráfica representa la cantidad de tweets y sus correspondientes _likes_, agrupado por día.

Cada punto representa un solo tweet, y su posición vertical o altura se alinea con la cantidad de _likes_ individualmente recibidos, la cual tiene un color indicado por la barra de colores a la derecha: mientras más rojo oscuro, más _likes_ individuales; mientras más amarillo claro, menos _likes_.

La cantidad de puntos por día es la cantidad de tweets hechos ese mismo día.

La línea azul muestra la suma de _likes_ de todos los tweets en un día.

La línea roja punteada marca el día 24 de marzo del 2021, fecha en la cual se publicó el video de la entrevista Clara Luz - Keith Raniere.

```python
from datetime import datetime

from data_processing import prepare_tweets_df, sum_likes_by_date
from plotter import plot_likes

TODAY = str(datetime.now().date())
tweets_clf = prepare_tweets_df(tweets_clf, before=TODAY)
likes = sum_likes_by_date(tweets_clf)

%matplotlib inline
plot_likes(likes)
```
